```python
from backtesting import Backtester
```

```python
# Initialize Backtester with AAPL stock
backtester = Backtester("AAPL", start="2020-01-01", end="2024-01-01")

# Fetch and display data
print("Fetching historical data...")
backtester.data.head()
```

## Visualizing Stock Price Data

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(backtester.data.index, backtester.data["Adj Close"], label="AAPL Price")
plt.title("Apple Stock Price (AAPL)")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

## Step 1: Running Moving Average Crossover


```python
print("Running Moving Average Crossover Strategy with Simulation...")
ma_results = backtester.moving_average_crossover()
print(ma_results)
```

## Step 2: Running RSI Strategy


```python
print("Running RSI Strategy with Simulation...")
rsi_results = backtester.rsi_strategy()
print(rsi_results)
```

##  Step 3: Running Bollinger Bands Strategy

```python
print("Running Bollinger Bands Strategy with Simulation...")
bb_results = backtester.bollinger_bands()
print(bb_results)
```
